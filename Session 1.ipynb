# Ejercicio 1.a

**Para los siguientes compuestos obtenga los diferentes tipos de representaci贸n molecular que se indican. Utilice el c贸digo visto en clase para obtener cada representaci贸n.**


```python
# Representaci贸n SMILES obtenida de ChemDraw
smiles = "O=C(NC1=C(C)C=CC=C1Cl)[C@H]2N(CCCC)CCCC2"

# Conversi贸n a formato mol
mol = Chem.MolFromSmiles(smiles)

# SMILES can贸nico
smiles_canonico = Chem.MolToSmiles(mol)

# Verificaci贸n de representaci贸n 2D (opcional)
#Draw.MolToImage(mol)

# InChI y InChIKey
inchi = Chem.MolToInchi(mol)
inchikey = Chem.MolToInchiKey(mol)

# Representaci贸n SMARTS
smarts = Chem.MolToSmarts(mol)

# Imprimir resultados
print(f'SMILES: {smiles}\n')
print(f'SMILES can贸nico: {smiles_canonico}\n')
print(f'SMARTS: {smarts}\n')
print(f'InChI: {inchi}\n')
print(f'InChIKey: {inchikey}\n')

# Representaci贸n 3d
ps = rdDistGeom.ETKDGv3()
cid = rdDistGeom.EmbedMolecule(mol,ps)
v = py3Dmol.view(width=300,height=300)
v.removeAllModels()
IPythonConsole.addMolToView(mol,v,confId=cid)
v.zoomTo()
print(f'Representaci贸n 3D:')
v.show()
```

```python
# URL (PubChem) morfina
morf_url = 'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/CID/5288826/record/SDF?record_type=3d&response_type=display'

# Conversi贸n a formato mol
morf_data = requests.get(morf_url).text
mol = Chem.MolFromMolBlock(morf_data)

# SMILES
smiles = Chem.MolToSmiles(mol)

# SMILES can贸nico
smiles_canonico = Chem.MolToSmiles(mol)

# InChI y InChIKey
inchi = Chem.MolToInchi(mol)
inchikey = Chem.MolToInchiKey(mol)

# Representaci贸n SMARTS
smarts = Chem.MolToSmarts(mol)

# Imprimir resultados
print(f'SMILES: {smiles}\n')
print(f'SMILES can贸nico: {smiles_canonico}\n')
print(f'SMARTS: {smarts}\n')
print(f'InChI: {inchi}\n')
print(f'InChIKey: {inchikey}\n')

# Representaci贸n 2d con estereoqu铆mica
IPythonConsole.drawOptions.addAtomIndices = False
IPythonConsole.drawOptions.addStereoAnnotation = True
mol2d = Chem.MolFromSmiles(smiles)

# Representaci贸n 3d
mol_with_hs = Chem.AddHs(mol) # A帽adir hidr贸genos expl铆citos
ps = rdDistGeom.ETKDGv3()
cid = rdDistGeom.EmbedMolecule(mol_with_hs,ps) # Usar la mol茅cula con hidr贸genos
mol3d = py3Dmol.view(width=300,height=300)
mol3d.removeAllModels()
IPythonConsole.addMolToView(mol_with_hs,mol3d,confId=cid) # Usar la mol茅cula con hidr贸genos
mol3d.zoomTo()
print(f'Representaci贸n 3D:')
mol3d.show()

print(f'Representaci贸n 2D con estereoqu铆mica:')
mol2d
```
